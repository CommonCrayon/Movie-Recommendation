```python
userbase1_dataframe = pd.read_csv("/content/drive/MyDrive/Movie Recommendation dataset/ml-100k/u1.base", names=['user_id', 'item_id', 'rating', 'timestamp'], delimiter="\t")
userbase1_dataframe = userbase1_dataframe.drop(["timestamp"], axis=1)
userbase1_dataframe
```

```python
user_list = userbase1_dataframe['user_id'].unique().tolist()
user_list
```

```python
usertest1_dataframe = pd.read_csv("/content/drive/MyDrive/Movie Recommendation dataset/ml-100k/u1.test", names=['user_id', 'item_id', 'rating', 'timestamp'], delimiter="\t")
usertest1_dataframe = usertest1_dataframe.drop(["timestamp"], axis=1)
usertest1_dataframe
```

```python
# Convert DataFrame to sparse matrix
user_item_matrix = userbase1_dataframe.pivot(index='user_id', columns='item_id', values='rating').fillna(0)
user_item_matrix_sparse = csr_matrix(user_item_matrix.values)

# Calculate cosine similarity matrix
cosine_sim_matrix = cosine_similarity(user_item_matrix_sparse)
cosine_sim_matrix
```

```python
utility_matrix = userbase1_dataframe.pivot(index='user_id', columns='item_id', values='rating').fillna(0)
utility_matrix
```

```python
# Convert cosine similarity matrix to DataFrame for better visualization
# Where both the rows and columns are labeled with user ids, and the values represent the cosine similarity between corresponding users based on their ratings.
cosine_sim_df = pd.DataFrame(cosine_sim_matrix, index=user_item_matrix.index, columns=user_item_matrix.index)
cosine_sim_df
```

```python
def rating_prediction_weighted(user_ids, cosine_sim_df, utility_matrix, top_n=10):
    all_predicted_ratings = []

    for user_id in user_ids:
        print(f"{user_id}/{len(user_ids)}")
        # Get cosine similarity scores for the given user
        user_similarity_scores = cosine_sim_df[user_id]

        # Sort users by similarity scores and get top N similar users (excluding the user itself)
        similar_users = user_similarity_scores.sort_values(ascending=False)[1:top_n+1]

        # Initialize a dictionary to store aggregated ratings from similar users
        predicted_ratings = {}

        # Aggregate ratings from similar users for items not yet rated by the given user
        unrated_items = utility_matrix.loc[user_id][utility_matrix.loc[user_id] == 0].index

        # Pre-calculate similarity scores for similar users
        similar_user_ids = similar_users.index
        similar_user_similarities = similar_users.values

        for item_id in unrated_items:
            # Get ratings of unrated items by similar users
            similar_user_ratings = utility_matrix.loc[similar_user_ids, item_id]

            # Calculate weighted sum and sum of weights
            weighted_sum = np.dot(similar_user_ratings, similar_user_similarities)
            sum_of_weights = np.sum(similar_user_similarities)

            # Calculate predicted rating if sum_of_weights is not zero
            if sum_of_weights != 0:
                predicted_rating = weighted_sum / sum_of_weights
                predicted_ratings[item_id] = predicted_rating

        # Append predicted ratings for the current user to the list
        all_predicted_ratings.extend([{'user_id': user_id, 'item_id': item_id, 'rating': rating} for item_id, rating in predicted_ratings.items()])

    return all_predicted_ratings
```

```python
users_id = user_list
predict_items_df = rating_prediction_weighted(users_id, cosine_sim_df, utility_matrix, 400)
```

```python
def evaluate_mse(predicted_ratings_df, test_df):
    # Merge predicted ratings with test DataFrame on user_id and item_id
    merged_df = pd.merge(test_df, predicted_ratings_df, on=['item_id'], how='inner')

    # Compute squared error between predicted and actual ratings
    merged_df['squared_error'] = (merged_df['rating'] - merged_df['predicted_rating']) ** 2

    # Calculate mean squared error (MSE)
    mse = merged_df['squared_error'].mean()

    return mse

# Example usage
```

```python
# Evaluation function
def evaluate_rmse(test_df, predicted_df):
    # Merge test and predicted DataFrames on item_id
    merged_df = test_df.merge(predicted_df, on='item_id', how='inner')

    # Compute Root Mean Squared Error (RMSE)
    rmse = np.sqrt(((merged_df['rating'] - merged_df['predicted_rating']) ** 2).mean())

    return rmse

```
